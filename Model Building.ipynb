# Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('survey_results_public.csv')
df.head()
```

## Selecting Necessary Columns


```python
df = df[['Country', 'EdLevel','YearsCodePro','Employment','ConvertedCompYearly']]
df = df.rename({'ConvertedCompYearly':'Salary'}, axis=1) #axis=1 for columns
df.head()
```

### Since we don't want to bother the user from gathering too much information for predicting salary, choosed only necessary columns.

```python
df = df[df['Salary'].notnull()]
df.head()
```

> Dropped records where the salary is null.

```python
df.info()
```

> * We have 48019 records and five features.
> * We have four object and 1 numerical feature.

```python
df = df.dropna() #we have lot of datapoints, so we dropped null records.
df.isnull().sum()
```

```python
df.shape
```

```python
df['Employment'].value_counts()
```

```python
df = df[df['Employment'] == 'Employed, full-time']
```

**Choosen only full time employeed software engineers for analysis and prediction purpose to provide to machine for pattern learning**

```python
df = df.drop('Employment',axis = 1)
df.info()
```

```python
df['Country'].value_counts()
```

**We don't want country with minimum patterns, so we're going to treat them**

```python
def shortern_categories(categories,cutoff):
    categorical_map={}
    for i in range(len(categories)):
        if categories.values[i] > cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shortern_categories(df.Country.value_counts(),400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize = (12, 7 ))
df.boxplot('Salary','Country',ax = ax) #box plot
plt.suptitle('Salary (US$) vs Country')
plt.title(' ')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

* As we see, there are few outliers in the data.
* We're going to treat them below

```python
df = df[df['Salary'] <= 250000]
df = df[df['Salary'] >=10000]
df = df[df['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1,1, figsize = (12, 7 ))
df.boxplot('Salary','Country',ax = ax) #box plot
plt.suptitle('Salary (US$) vs Country')
plt.title(' ')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df['YearsCodePro'].unique()
```

```python
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

#applying user-defined function:
df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(x):
    if "Bachelors degree" in x:
        return "Bachelor's degree"
    if "Masters degree" in x:
        return "Master's degree"
    if 'Professional degree' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df['EdLevel'].unique()
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder
le_education= LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df['EdLevel'].unique()
#le.classes_
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df['Country'].unique()
```

## Training

```python
X = df.drop('Salary', axis = 1)
y = df['Salary']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X,y.values) #input all data for training into the model
```

```python
y_pred = lr_model.predict(X)
```

```python
from sklearn.metrics import  mean_squared_error,mean_absolute_error,r2_score
```

```python
rmse_error = np.sqrt(mean_squared_error(y,y_pred))
print("${:,.02f}".format(rmse_error))
print('R2 Score:',r2_score(y,y_pred))
```

    Error is very high

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=0)
dt_model.fit(X,y.values)
y_pred_dt = dt_model.predict(X)
```

```python
rmse_error_dt = np.sqrt(mean_squared_error(y,y_pred_dt))
print("${:,.02f}".format(rmse_error_dt))
print('R2 Score:',r2_score(y,y_pred_dt))
```

    Slightly reduced error

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=0)
rf_model.fit(X,y.values)
y_pred_rf = rf_model.predict(X)
```

```python
rmse_error_rf = np.sqrt(mean_squared_error(y,y_pred_rf))
print("${:,.02f}".format(rmse_error_rf))
print('R2 Score:',r2_score(y,y_pred_rf))
```

    error is not satisfied with this model too

```python
from sklearn.model_selection import GridSearchCV

params = {'max_depth': [None,2,4,6,8,10,12]}
dt = DecisionTreeRegressor(random_state=10)
gs = GridSearchCV(estimator=dt,param_grid=params, scoring='neg_mean_squared_error')
gs.fit(X,y.values)
```

```python
gs.best_params_
```

```python
regressor = gs.best_estimator_
regressor.fit(X,y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
r2_score(y,y_pred)
```

# User-Defined function for trying all regression models at a time 

```python
perf_score = pd.DataFrame(columns=["Model", 'RMSE', 'R-squared'])

def update_performance(name, model, Y, pred):
    global perf_score
    perf_score= perf_score.append({
        'Model': name,
        'RMSE': "${:,.02f}".format(np.sqrt(mean_squared_error(Y, pred))),
        'R-squared': r2_score(Y, pred)
    }, ignore_index=True)

```

```python
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
import warnings
warnings.filterwarnings('ignore')
model = {
    "LinearRegression": LinearRegression(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "AdaboostRegressor": AdaBoostRegressor(),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "XGBoostRegressor": XGBRegressor(),
}

for clf_name, clf in model.items():
    clf.fit(X, y.values)
    ypred = clf.predict(X)
    update_performance(clf_name, clf, y, ypred)

perf_score
```
